```python
from bs4 import BeautifulSoup

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
lines=soup.find_all('a',class_="popwin-ss")
```

```python
new=[]
for line in range(8,len(lines)):
  new.append(lines[line].text)
```

```python
new[0]
```

```python
l=soup.find('ul',class_="exhibitor").children
```

```python
l
```

```python
links=[]
for i in l:
  links.append(i)
```

```python
links[1].children
```

```python
faltu=[]
for i in links:
  if i!='\n':
    child=[]
    for j in i.children:
      child.append(j.text)
    faltu.append(child)
```

```python
faltu
```

```python
import pandas as pd
```

```python
final=pd.DataFrame(faltu,columns=['Online','Onsite','Booth No','Show Name','Company Name'])
```

```python
final.to_csv('faltu.csv')
```
