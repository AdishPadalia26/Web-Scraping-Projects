```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
response=requests.get('https://studyabroad.shiksha.com/usa/universities-1',headers=headers).text
```

```python
soup=BeautifulSoup(response,'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('a',class_='font-15'):
    print(i.text)
```
